```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df= pd.read_csv(r'D:\3rd Semester DATA SCINCE\preprocessed_data.csv')
```

```python
sentiment_counts = df["scoreSentiment"].value_counts()
print(sentiment_counts)
```

```python
df["reviewLength"] = df["reviewText"].apply(len)
print(df["reviewLength"])
```

```python
descriptive_stats = {
    "Sentiment Distribution": sentiment_counts.to_dict(),
    "Review Length Stats": df["reviewLength"].describe().to_dict(),
}
descriptive_stats
```

```python
plt.figure(figsize=(4,5))
sns.countplot(x="scoreSentiment", data=df, palette="viridis")
plt.title("Sentiment Distribution", fontsize=16)
plt.xlabel("Sentiment (0 = Negative, 1 = Positive)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Visualize review length distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["reviewLength"], bins=50, kde=True, color="blue")
plt.title("Review Length Distribution", fontsize=16)
plt.xlabel("Review Length", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
pos = (df['sentiment_score'] > 0).sum()
neg = (df['sentiment_score'] < 0).sum()
neu = (df['sentiment_score'] == 0).sum()

plt.pie([pos, neg, neu], labels=['Positive', 'Negative', 'Neutral'], autopct='%1.1f%%')
plt.title('Sentiment Analysis')
plt.show()
```
